#### Sentence Transformers on Hugging Face

- Hugging Face sentence-transformers is a python framework for state-of-the-art sentence, text and image embeddings.
- One of the embedding models is uded in the HuggingFaceEmbeddings class
- We also added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model='all-MiniLM-L6-v2')
```

```python
text = 'This is a test document'

qr = embedding_model.embed_query(text)
qr
```

```python
len(qr)
```

```python
text2 = 'This is a test document'

dr = embedding_model.embed_documents([text, text2])
dr
```

```python
len(dr[0])
```
